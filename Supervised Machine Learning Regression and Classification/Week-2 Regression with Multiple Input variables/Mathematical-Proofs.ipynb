## 1. Probability

### Intuition
You wake up, sky is cloudy. You think — *"maybe 70% chance it rains."*
That's it. Probability is just your **belief about how likely something is to happen.**
Expressed as a number between 0 and 1.
- 0 = impossible
- 1 = certain
- 0.7 = probably but not sure

### Why do we need it?
So we can **predict something** based on every possible outcome — how likely is OUR outcome going to happen?

### Formula
$$P(event) = \frac{\text{ways that event can happen}}{\text{total possible outcomes}}$$

### Example
Flip a coin. What's P(heads)?
- Ways heads can happen = 1
- Total outcomes = 2
$$P(heads) = \frac{1}{2} = 0.5$$

### Proof / Derivation
No deep derivation needed here. It's an **axiom** — a starting definition.
But it must satisfy 3 rules (Kolmogorov Axioms):
1. $P(A) \geq 0$ — probability is never negative
2. $P(\text{entire sample space}) = 1$ — something must happen
3. $P(A \cup B) = P(A) + P(B)$ if A and B can't happen together

## 2. Discrete vs Continuous Distributions

### The Key Question to Ask Yourself
> *"Can I COUNT the outcomes... or do they FLOW continuously?"*

### Discrete — You Can Count
- Roll a dice → 1, 2, 3, 4, 5, 6. Finite. Nothing in between. Can't roll 3.7.
- Number of bedrooms → 1, 2, 3, 4. Countable.
- Star rating of a call → 1 to 5. Countable.

**Discrete = countable, separated, distinct values.**

### Continuous — It Flows
- Height of a person → 5.7? 5.71? 5.711? Never stops. Infinite values between any two numbers.
- Waiting time for a bus → could be 4.99 mins, 5.001 mins, 5.0001 mins...
- House size → flows forever at decimal level.

**Continuous = infinite possible values. Flows like water. Fingers aren't enough to count.**

### One Line
> **Discrete = count it on fingers. Continuous = flows like water.**

### Probability for Continuous?
For continuous data, probability of any EXACT value = 0 (infinite possibilities).
So instead we ask: probability of falling in a **range**.
$$P(a \leq x \leq b) = \int_a^b f(x)\,dx$$
where $f(x)$ is the **probability density function (PDF)**.

## 3. Variance

### Intuition
Two groups of students, both with mean = 50:
- Group 1: 48, 49, 50, 51, 52 → all close together, low spread
- Group 2: 5, 10, 50, 90, 95 → flying everywhere, high spread

Same mean. But something is very different. That difference is **variance.**
> **Variance = how spread out is your data from the center?**

### How Would You Invent This?
Step 1 — find distance of each value from mean:
$(x_i - \mu)$

Step 2 — add them up... but positives and negatives cancel → sum = 0. Useless.

Step 3 — kill the negatives by **squaring** each distance:
$(x_i - \mu)^2$

Step 4 — average them (so it's per person, not total):
$$\sigma^2 = \frac{1}{n}\sum_{i=1}^{n}(x_i - \mu)^2$$

### Formula
$$\boxed{\sigma^2 = \frac{1}{n}\sum_{i=1}^{n}(x_i - \mu)^2}$$

### Example
Group 1: [48, 49, 50, 51, 52], mean = 50
$$\sigma^2 = \frac{(48-50)^2+(49-50)^2+(50-50)^2+(51-50)^2+(52-50)^2}{5} = \frac{4+1+0+1+4}{5} = 2$$

---
## 4. Standard Deviation

### Intuition
Variance is in **squared units** (score²). That doesn't mean anything to a human brain.
So we just **undo the squaring** → take the square root → back to original units.

> **Standard Deviation = variance brought back to real units. Average distance from the mean.**

### How Would You Invent This?
You invented variance. But then you realized — units are squared. Weird.
Simple fix: $\sqrt{\text{variance}}$

### Formula
$$\boxed{\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i - \mu)^2}}$$

### One Line
> **Variance = spread in squared units. Standard Deviation = same thing, back to reality.**

## 5. Covariance

### Intuition
Variance watches ONE feature's spread.
But what if you want to know — **do two features move together?**
- When house SIZE goes up, does house PRICE also go up?
- When you eat more, does weight also go up?

> **Covariance = measures how two features move together.**
> *"Secretary of Variance — watches two people instead of one."* 😄

### How Would You Invent This?
Variance used: $(x_i - \mu_x)^2$ → squared ONE feature's distance.
For two features, instead of squaring — **multiply both distances together:**
$(x_i - \mu_x)(y_i - \mu_y)$

- Both go up → both positive → product positive → **positive covariance** ✅
- One up, one down → one positive, one negative → product negative → **negative covariance** ✅
- No relationship → sometimes +, sometimes − → cancels → **zero covariance** ✅

### Formula
$$\boxed{\text{Cov}(x,y) = \frac{1}{n}\sum_{i=1}^{n}(x_i - \mu_x)(y_i - \mu_y)}$$

### No Deep Derivation Needed!
This is English → Math translation only. Someone said:
> *"I want to measure how two things move together... distance from center... multiply them... average it."*
That's it. ✅

## 6. Distribution

### Intuition
Ask: *"In a class of 100 students, how many got each score?"*
- 10 students got 40
- 20 students got 50
- 40 students got 60
- 20 students got 70
- 10 students got 80

That **pattern of how values are spread across possibilities** — that's a distribution.

> **Distribution = the pattern of how your data is spread. Who lives where.**

## 7. Normal / Gaussian Distribution

### Intuition
Draw the student score distribution from above. What shape do you get?

        *
       ***
      *****
     *******
    *********
40  50  60  70  80

Tall in the middle. Short on the edges. **A bell. 🔔**

### Why Does Nature Love This Shape?
Heights, weights, exam scores, measurement errors — most things are average.
Very few are extreme. So when you plot real world data, you almost always get a bell.

**The inventor (Gauss) didn't invent it — he just named what he kept seeing in data.**
We studied it in reverse: name first, then understood why.

### Two Numbers Completely Describe the Bell
- **Mean (μ)** → where is the CENTER of the bell?
- **Standard Deviation (σ)** → how FAT or THIN is the bell?
  - Wide bell = high σ = data spread far
  - Thin bell = low σ = data packed close

most data distributed to center

Okay. So the formula needs to output a high value when $x$ is close to the mean $\mu$, and a low value when $x$ is far from $\mu$.

What simple mathematical expression gives you something HIGH when $x$ is close to $\mu$ and LOW when $x$ is far?

When we did softmax — we used $e^z$ remember?

We said $e^z$ is always positive and when $z$ is large $e^z$ is large, when $z$ is small $e^z$ is small.

Now think about this:

What if $z$ was negative distance from mean?

$x$ is very close to $\mu$ → distance is small → $-\text{distance}$ is small negative → $e^{\text{small negative}} \approx 1$ (HIGH) ✅

$x$ is very far from $\mu$ → distance is large → $-\text{distance}$ is large negative → $e^{\text{large negative}} \approx 0$ (LOW) ✅

So the core of normal distribution is just:

$$
e^{-\text{something related to distance}}
$$

So putting it all together:

$$
\text{distance} = \frac{(x - \mu)^2}{\sigma^2}
$$

Plug into $e$:

$$
e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

The 2 in denominator comes from calculus — makes derivative clean. Same reason we had 2 in gradient descent! ✅

Now one last piece — this gives us shape but not a valid probability.

Why? Because probabilities must integrate to 1 over all $x$.

So we divide by a normalizing constant:

$$
p(x) = \frac{1}{\sqrt{2\pi}\sigma} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

The $\frac{1}{\sqrt{2\pi}\sigma}$ is just a scaling constant that forces the total area under curve = 1. ✅

### Formula (PDF)
$$
\boxed{f(x) = \frac{1}{\sigma\sqrt{2\pi}}\, e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}}
$$

### Breaking Down the Formula
- $\frac{x - \mu}{\sigma}$ → how far are you from center, in standard deviation units
- The exponent is negative → farther from center = smaller value = less likely ✅
- $e^{-\text{something}}$ → always between 0 and 1, creates the bell curve shape
- $\frac{1}{\sigma\sqrt{2\pi}}$ → scaling factor so total area under curve = 1 (total probability = 1)

### Normal vs Gaussian?
> **Same thing. Different name. Gauss was the mathematician who formalized it.**

### Can Normal Distribution be Discrete?
> Technically No. Normal distribution is always continuous.
> Discrete data can form bell shapes (Binomial distribution) but it's a different name, same soul.

## 8. Z-Score Normalization

### The Original Problem
Feature x1 (house size): 200 → 600
Feature x2 (bedrooms): 0 → 5

Different scales → gradient descent bounces around inefficiently.
**We need to bring everything to a comparable small range.**

### How Would You Invent Z-Score?

**Step 1** — Center the data (remove the mean):
$(x - \mu)$ → now data lives around 0. Centered. ✅

*Why centering?* Because outliers in raw data crush everything near zero when you just divide by max.
Centering first makes the data balanced.

**Step 2** — Scale by true spread (divide by std deviation):
$\frac{x - \mu}{\sigma}$ → now spread is normalized to 1 unit. ✅

*Why σ?* Because σ is the TRUE spread of your data — not affected by outliers like max is.

### Formula
$$\boxed{z = \frac{x - \mu}{\sigma}}$$

### What Does Each Part Mean?
- **Numerator** $(x - \mu)$: How far are you from the center?
- **Denominator** $(\sigma)$: What is the true spread of the whole data?
- **Result**: What fraction of the total spread are you away from center?

### Result After Z-Score
- Mean of transformed data = **0** always
- Standard deviation of transformed data = **1** always
- Data lives roughly in **-3 to +3** (within 3 standard deviations)

### Why Is This Better Than Just Dividing by Max?
| Method | Formula | Problem |
|--------|---------|----------|
| Divide by max | $x / x_{max}$ | One outlier ruins everything |
| Min-max scaling | $(x - x_{min})/(x_{max} - x_{min})$ | Still sensitive to outliers |
| Z-score | $(x - \mu)/\sigma$ | Robust, centered, uses true spread ✅ |

### Connection to Normal Distribution
When you apply Z-score to normally distributed data, you get the **Standard Normal Distribution**:
$$\mathcal{N}(0, 1) \text{ — mean=0, std=1}$$
This is why the formula inside the Normal Distribution PDF was $\frac{x-\mu}{\sigma}$ — **it's the Z-score!**
The entire chain connects here. 🔥

### One Line to Remember Forever
> **Z-score asks: 'For this value, how far from center am I, expressed as a fraction of the true spread?'**
