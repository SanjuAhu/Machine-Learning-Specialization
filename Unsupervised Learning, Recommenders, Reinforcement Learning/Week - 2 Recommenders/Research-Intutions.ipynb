The History of the Recommender (The Billion-Dollar Math)
You asked who invented the math that builds billionaires and, as you correctly noted, has a dark side that exploits human psychology. It didn't start with TikTok or YouTube. It started with academics trying to solve a simple problem: too much information.

1. The Academics (1994): Paul Resnick and John Riedl
Before the modern web, people used "Usenet" to read text articles. There was too much noise. Resnick and Riedl at the University of Minnesota built a system called GroupLens. They invented the concept of Collaborative Filtering.
The Intuition: If Sanjay and I both rated the exact same 10 articles highly, our weights are mathematically aligned. Therefore, if I read an 11th article and like it, the system should recommend it to Sanjay. They built it to help humanity find signal in the noise.

2. The E-Commerce Hacker (1998): Greg Linden
Amazon was a tiny company selling books. Greg Linden was an engineer there. He realized that doing the GroupLens math (comparing millions of users to each other in real-time) was too computationally heavy; the servers would crash.
The Intuition: Linden flipped the matrix. Instead of finding similar users, he found similar items. He wrote the algorithm for "Customers who bought this item also bought..." Amazon executives actually ordered him to stop working on it because they thought it would distract people from the checkout page. Linden secretly ran an A/B test anyway. It generated so much revenue that Jeff Bezos made it the core of the entire company.

3. The Netflix Prize and The Matrix (2006): Simon Funk
This is the moment the math became lethal. Netflix offered $1,000,000 to anyone who could improve their recommendation algorithm by 10%.
An independent researcher named Simon Funk (his real name is Webb) realized that the data was a giant, mostly empty grid of users and movies. He used a linear algebra technique to compress the matrix.
The Intuition: He proved mathematically that you do not need to know the features of the movie. The algorithm doesn't need to know if a movie is "Action" or "Romance." If you run Gradient Descent on the empty grid, the math will automatically invent its own hidden features based purely on human behavior.

The Notation Engine:

The Prediction:

If user $j$ has not eaten food $i$, we predict their rating using the exact same line equation from Course 1:$$\text{Prediction} = w^{(j)} \cdot x^{(i)} + b^{(j)}$$

The Global Cost Function:You correctly identified that we just sum up the Linear Regression cost for all users, but only for the foods they actually rated (where $r(i,j) = 1$).

$$J(w, b) = \frac{1}{2} \sum_{j=1}^{n_u} \sum_{i: r(i,j)=1} (w^{(j)} \cdot x^{(i)} + b^{(j)} - y^{(i, j)})^2 + \text{Regularization}$$

(Note: You correctly noticed that Andrew Ng drops the $\frac{1}{m^{(j)}}$ term here. In recommender systems, users rate drastically different numbers of items, so removing the $m$ makes the math much cleaner without breaking the optimization).

🚨 Resolving Your Doubt: How Regularization Actually PenalizesYou asked: "I forget that how does the mathematically regularization penalize that?"Let's look at the pure physics of the math.The algorithm's entire goal in life is to make the Cost Function ($J$) as small as possible (close to 0).The Regularization term looks like this: $+ \frac{\lambda}{2} (w)^2$Imagine the algorithm finds a way to perfectly memorize a user's ratings by making the weight incredibly large, like $w = 1000$.If it does that, the prediction error part of the equation becomes $0$. But look at what happens to the Regularization part:$+ \frac{\lambda}{2} (1000)^2 = + 500,000$.

Suddenly, the total Cost ($J$) explodes to $500,000$! The algorithm panics. It realizes, "If I make the weights too big, the overall cost gets destroyed." Therefore, mathematically, it is forced to keep the weights ($w$) as small as possible while still trying to fit the data. Small weights mean a smoother, simpler line that doesn't overfit to just a few weird ratings. That is how the math physically penalizes complexity.

DERIVING COLLABORATIVE FILTERINGHere is our global, massive Cost Function for Collaborative Filtering:$$J(w, b, x) = \frac{1}{2} \sum_{(i,j): r(i,j)=1} (w^{(j)} \cdot x^{(i)} + b^{(j)} - y^{(i,j)})^2 + \frac{\lambda}{2} \sum_{j=1}^{n_u} (w^{(j)})^2 + \frac{\lambda}{2} \sum_{i=1}^{n_m} (x^{(i)})^2$$Let's make the math visually cleaner. Let's define the Error for a specific user $j$ and movie $i$ as:$$\text{Error}^{(i,j)} = (w^{(j)} \cdot x^{(i)} + b^{(j)}) - y^{(i,j)}$$Now, to run Gradient Descent, we need to find three partial derivatives: $\frac{\partial J}{\partial w}$, $\frac{\partial J}{\partial x}$, and $\frac{\partial J}{\partial b}$.

1. The Derivative with respect to $x^{(i)}$ (Learning the Movie's Features)We want to know how to adjust the features of a specific movie $i$. We use the Chain Rule from calculus.
- The $2$ from the squared exponent drops down to the front.
- It cancels out the $\frac{1}{2}$.
- We leave the inside of the parenthesis alone (which is our $\text{Error}^{(i,j)}$).
- We multiply by the derivative of the inside with respect to $x^{(i)}$. Since $w^{(j)} \cdot x^{(i)}$ is just a constant times $x$, the derivative is just $w^{(j)}$.
- Finally, we take the derivative of the regularization term: the $2$ drops down, cancels the $\frac{1}{2}$, leaving $\lambda x^{(i)}$.

The Final Equation for $x$:$$\frac{\partial J}{\partial x^{(i)}} = \sum_{j: r(i,j)=1} \left( \text{Error}^{(i,j)} \right) w^{(j)} + \lambda x^{(i)}$$(In plain English: To update a movie's features, look at every user $j$ who rated it, multiply their error by their personal weight $w^{(j)}$, add it all up, and add the regularization penalty).

2. The Derivative with respect to $w^{(j)}$ (Learning the User's Taste)Because the equation is perfectly symmetrical, the calculus is the exact same, just mirrored!We take the derivative of the inside with respect to $w^{(j)}$, which leaves $x^{(i)}$.The Final Equation for $w$:$$\frac{\partial J}{\partial w^{(j)}} = \sum_{i: r(i,j)=1} \left( \text{Error}^{(i,j)} \right) x^{(i)} + \lambda w^{(j)}$$(In plain English: To update a user's taste profile, look at every movie $i$ they rated, multiply the error by the movie's features $x^{(i)}$, add it all up, and add the penalty).

3. The Derivative with respect to $b^{(j)}$ (The User's Baseline)The bias term $b^{(j)}$ has no $x$ or $w$ attached to it. The derivative of $b^{(j)}$ by itself is just $1$. Also, notice in the global cost function, we do not regularize $b$.The Final Equation for $b$:$$\frac{\partial J}{\partial b^{(j)}} = \sum_{i: r(i,j)=1} \left( \text{Error}^{(i,j)} \right)$$

The Beauty of the CalculusLook at those three derived equations.When your Python code runs w = w - alpha * dw, it is running these exact three equations millions of times a second.You start with completely random garbage numbers for $w$ and $x$.The machine looks at the first equation, pushes $x$ a tiny bit to match the user's weights.Then it looks at the second equation, pushes $w$ a tiny bit to match the new movie features.It loops this over and over.Because of the calculus you just derived, the math physically forces the random numbers to sculpt themselves into the true hidden features of the dataset.
