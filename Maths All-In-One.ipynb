# Regression (Continuous Mapping)
- The Hypothesis ($h_{\mathbf{w},b}(\mathbf{x})$): Your proposed model.$$h_{\mathbf{w},b}(\mathbf{x}) = w_1x_1 + w_2x_2 + \dots + w_nx_n + b$$

- The Cost Function ($J(\mathbf{w}, b)$): The measure of your failure. In Linear Regression, we use Mean Squared Error (MSE):$$J(\mathbf{w},b) = \frac{1}{2m} \sum_{i=1}^{m} (h_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)})^2$$

Cost Function:$$J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} \left( (wx^{(i)} + b) - y^{(i)} \right)^2$$

What are the final, simplified equations for $\frac{\partial J}{\partial w}$ and $\frac{\partial J}{\partial b}$?

The Weight Gradient:$$\frac{\partial J}{\partial w} = \frac{1}{2m} \sum_{i=1}^{m} 2 \left( f_{w,b}(x^{(i)}) - y^{(i)} \right) x^{(i)} = \frac{1}{m} \sum_{i=1}^{m} \left( f_{w,b}(x^{(i)}) - y^{(i)} \right) x^{(i)}$$

The Bias Gradient:$$\frac{\partial J}{\partial b} = \frac{1}{2m} \sum_{i=1}^{m} 2 \left( f_{w,b}(x^{(i)}) - y^{(i)} \right) (1) = \frac{1}{m} \sum_{i=1}^{m} \left( f_{w,b}(x^{(i)}) - y^{(i)} \right)$$


The Formalization:
$$
w := w - \alpha \frac{\partial J(w,b)}{\partial w}
$$

$$
b := b - \alpha \frac{\partial J(w,b)}{\partial b}
$$

# MULTIPLE REGRESSION & VECTORIZATION

The Formalization (Vectorized):$$f_{\mathbf{w},b}(\mathbf{x}) = \mathbf{w} \cdot \mathbf{x} + b$$

Gradient -  w $$\mathbf{w} := \mathbf{w} - \alpha \frac{1}{m} \mathbf{X}^T (\mathbf{Xw} + b - \mathbf{y})$$

Bias - b $$\mathbf{b} := \mathbf{b} - \alpha \frac{1}{m} \sum_{i=1}^{m} (\mathbf{Xw} + b - \mathbf{y})$$

The Max Scaling: $x_{scaled} = \frac{x}{max}$

The Mean Normalization: $x_{scaled} = \frac{x - \mu}{max - min}$

The Z-Score normalization: $$x_j^{(i)} := \frac{x_j^{(i)} - \mu_j}{\sigma_j}$$

Feature Engineering (Polynomial Regression) $$f_{\mathbf{w},b}(x) = w_1x + w_2x^2 + w_3x^3 + b$$
